## Question 2: What genre of movie is each director best known for?

```python
cur = conn.cursor()
cur.execute("""SELECT * 
                FROM imdb_title_ratings_gz 
                LIMIT 5;""").fetchall()
```

## 2.1 What is Top 1 movie for each director? 

```python
cur.execute(""" SELECT directors, tconst, 
                MAX(averagerating) AS max_rating
                FROM imdb_title_crew_gz
                JOIN imdb_title_ratings_gz
                USING(tconst)
                WHERE directors IS NOT NULL
                GROUP BY directors, tconst
                ORDER BY max_rating DESC;""")
directors_df = pd.DataFrame(cur.fetchall())
directors_df.columns = [x[0] for x in cur.description]
directors_df.head(10)
```

```python
cur.execute(""" SELECT directors, 
                COUNT(directors) AS num_of_movies,
                MAX(averagerating) AS highest_rating,
                MIN(averagerating) AS lowest_rating,
                AVG(averagerating) AS mean_rating,
                SUM(averagerating) AS total_rating
                FROM imdb_title_crew_gz
                JOIN imdb_title_ratings_gz
                USING(tconst)
                GROUP BY directors
                HAVING num_of_movies > 0
                ORDER BY total_rating DESC;""")
directors_df = pd.DataFrame(cur.fetchall())
directors_df.columns = [x[0] for x in cur.description]
directors_df.head(100)
```

## 2.2 What genre of movie is each director best known for?

```python
cur.execute(""" SELECT directors, tconst, genres, 
                MAX(averagerating) AS max_rating
                FROM imdb_title_crew_gz
                JOIN imdb_title_ratings_gz
                USING(tconst)
                JOIN imdb_title_basics_gz
                USING (tconst)
                WHERE directors IS NOT NULL
                GROUP BY directors, tconst
                ORDER BY max_rating DESC;""")
directors_df = pd.DataFrame(cur.fetchall())
directors_df.columns = [x[0] for x in cur.description]
directors_df.head(10)
```

## 2.3 What are Top 5 movie genres that most directors are known for in various rating ranges, e.g. 9.0-10.0, 8.0-9.0, 7.0-8.0, and so on?

```python
cur.execute(""" SELECT directors, tconst, genres, 
                MAX(averagerating) AS max_rating
                FROM imdb_title_crew_gz
                JOIN imdb_title_ratings_gz
                USING(tconst)
                JOIN imdb_title_basics_gz
                USING (tconst)
                WHERE directors IS NOT NULL
                GROUP BY directors, tconst
                HAVING max_rating BETWEEN 9 AND 10
                ORDER BY max_rating DESC;""")
directors_df = pd.DataFrame(cur.fetchall())
directors_df.columns = [x[0] for x in cur.description]
print(len(directors_df))
print(directors_df.head(500))
directors_df.tail(500)
```

```python
cur.execute(""" SELECT directors, tconst, genres, 
                MAX(averagerating) AS max_rating
                FROM imdb_title_crew_gz
                JOIN imdb_title_ratings_gz
                USING(tconst)
                JOIN imdb_title_basics_gz
                USING (tconst)
                WHERE directors IS NOT NULL
                GROUP BY directors, tconst
                HAVING max_rating >= 8 AND max_rating < 9
                ORDER BY max_rating DESC;""")
directors_df = pd.DataFrame(cur.fetchall())
directors_df.columns = [x[0] for x in cur.description]
print(len(directors_df))
print(directors_df.head(500))
directors_df.tail(500)
```
